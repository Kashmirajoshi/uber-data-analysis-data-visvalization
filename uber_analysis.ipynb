```python

data = {
    'CATEGORY*': ['Business', 'Personal'],
    'MILES*': [11487.0, 717.7]
}
uber_df = pd.DataFrame(data)

# Step 3: Group by CATEGORY and sum the MILES
df = uber_df.groupby(['CATEGORY*']).sum()

# Step 4: Calculate the proportions
total_miles = df['MILES*'].sum()  # Total miles
Business = df.loc['Business', 'MILES*'] / total_miles if total_miles > 0 else 0
Personal = df.loc['Personal', 'MILES*'] / total_miles if total_miles > 0 else 0

# Step 5: Print the results
print("Proportion of Business :", Business)
print("Proportion of Personal :", Personal)
```
